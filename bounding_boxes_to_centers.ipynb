```python
#Using Pandas, we transform the YOLO prescription for bounding box labels to single-pixel centers.
#Original data: One line for each bounding box for each image id, with its bounding box coordinates in the 
#form [xmin, ymin, width, height].
#Transformed data: Single-pixel centers. For each image, a line in the form [filename, counts, locations]

import pandas as pd
```

```python
df = pd.read_csv('train.csv')
df.drop('width', inplace=True, axis=1)
df.drop('height', inplace=True, axis=1)
df.drop('source', inplace=True, axis=1)
```

```python
df['bbox'] = df['bbox'].str[1:-1]
df['bbox'] = df['bbox'].apply(lambda x : x.split(','))

listx1 = []
listx2 = []
listy1 = []
listy2 = []

for i, row in df['bbox'].iteritems():
    listx1.append(row[0])
    listx2.append(row[1])
    listy1.append(row[2])
    listy2.append(row[3])

df['xmin'] = listx1
df['ymin'] = listx2
df['width'] = listy1
df['height'] = listy2

df.drop('bbox', inplace=True, axis=1)

df['xcenter'] = (df['xmin'].astype(float) + (df['width'].astype(float)) / 2).astype(int)
df['ycenter'] = (df['ymin'].astype(float) + (df['height'].astype(float)) / 2).astype(int)

df.drop('xmin', inplace=True, axis=1)
df.drop('ymin', inplace=True, axis=1)
df.drop('width', inplace=True, axis=1)
df.drop('height', inplace=True, axis=1)

df['center'] = '(' + df['xcenter'].astype(str) + ", " + df['ycenter'].astype(str) + ")"

df.drop('xcenter', inplace=True, axis=1)
df.drop('ycenter', inplace=True, axis=1)

df = df.groupby("image_id")["center"].apply(list).reset_index(name='locations')

df['count'] = df['locations'].str.len()

df = df [['image_id', 'count', 'locations']]

df.columns = ['filename', 'count', 'locations']

df['filename'] = df['filename'].astype(str) + '.jpg'

print(df.sort_values(by=['count'], ascending=True).head(50))
```
